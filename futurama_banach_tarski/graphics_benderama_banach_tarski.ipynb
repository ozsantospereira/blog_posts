```python
import numpy as np
import matplotlib.pyplot as plt

# Gráfico 1: Crescimento da Soma Harmônica
n = np.arange(1, 101)
H_n = np.cumsum(1 / n)

fig1 = plt.figure(figsize=(8, 5))
plt.plot(n, H_n, marker='o')
plt.title('Crescimento da Soma Harmônica \n $H_n = \\sum_{k=1}^n \\frac{1}{k}$')
plt.xlabel('n')
plt.ylabel('$H_n$')
plt.grid(True)
plt.tight_layout()

# Gráfico 2: Número de Benders ao longo das iterações
k = np.arange(0, 25)
num_benders = 2 ** k
massa_individual = 1 / (2 ** k * (k + 1))

fig2 = plt.figure(figsize=(8, 5))
plt.plot(k, num_benders, label='$2^k$: Número de Benders', marker='o')
plt.title('Duplicação de Benders por iteração \n $2^k$')
plt.xlabel('Iteração $k$')
plt.ylabel('Número de Benders')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Gráfico 3: Massa individual por Bender
fig3 = plt.figure(figsize=(8, 5))
plt.plot(k, massa_individual, label=r'$M_k = \frac{M_0}{2^k(k+1)}$', marker='x')
plt.title('Massa individual de cada Bender \n $M_k = \\frac{M_0}{2^k(k+1)}$')
plt.xlabel('Iteração $k$')
plt.ylabel('Massa Individual')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Gráfico 4: Massa Total acumulada ao longo das iterações
massa_total = np.cumsum(num_benders * massa_individual)

fig4 = plt.figure(figsize=(8, 5))
plt.plot(k, massa_total, label='Massa Total acumulada', marker='s')
plt.title('Massa Total acumulada \n $M = \\sum_{k=0}^n \\frac{M_0}{k+1}$')
plt.xlabel('Iteração $k$')
plt.ylabel('Massa Total')
plt.grid(True)
plt.legend()
plt.tight_layout()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Recriação da árvore com layout manual
G = nx.DiGraph()

# Criando a árvore até k=3 gerações
G.add_edges_from([
    ('B0', 'B1'), ('B0', 'B2'),
    ('B1', 'B3'), ('B1', 'B4'),
    ('B2', 'B5'), ('B2', 'B6'),
    ('B3', 'B7'), ('B3', 'B8'),
    ('B4', 'B9'), ('B4', 'B10'),
    ('B5', 'B11'), ('B5', 'B12'),
    ('B6', 'B13'), ('B6', 'B14')
])

# Layout manual simulando estrutura de árvore
pos = {
    'B0': (0, 0),
    'B1': (-2, -1), 'B2': (2, -1),
    'B3': (-3, -2), 'B4': (-1, -2), 'B5': (1, -2), 'B6': (3, -2),
    'B7': (-3.5, -3), 'B8': (-2.5, -3),
    'B9': (-1.5, -3), 'B10': (-0.5, -3),
    'B11': (0.5, -3), 'B12': (1.5, -3),
    'B13': (2.5, -3), 'B14': (3.5, -3)
}

fig_tree_fixed = plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=800, node_color="lightgreen", arrows=False)
plt.title("Árvore de Duplicação dos Benders até k=3")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Limitar n para manter tudo visual sem overflow
n = np.arange(1, 31)

# Séries ajustadas
harmonica = np.cumsum(1 / n)
serie_p09 = np.cumsum(1 / n**0.5)
serie_constante = np.cumsum(np.ones_like(n))
serie_linear = np.cumsum(n)  # Nova série: soma de n
serie_exp = np.cumsum(2 ** n)  # Exponencial até n=30

# Gráfico comparativo ajustado
plt.figure(figsize=(10, 6))
plt.plot(n, harmonica, label='Série Harmônica $\\sum \\frac{1}{n}$')
plt.plot(n, serie_p09, label='$\\sum \\frac{1}{n^{0.5}}$')
plt.plot(n, serie_constante, label='$\\sum 1$')
plt.plot(n, serie_linear, label='$\\sum n$')
plt.plot(n, serie_exp, label='$\\sum 2^n$')
plt.yscale('log')  # Escala logarítmica para comparar
plt.title('Comparação de Séries Divergentes (ajustada)')
plt.xlabel('n')
plt.ylabel('Soma acumulada (escala log)')
plt.legend()
plt.grid(True, which='both', linestyle='--')
plt.tight_layout()
plt.show()
```
